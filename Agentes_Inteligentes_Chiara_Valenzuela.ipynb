```
from IPython.display import display
import ipywidgets as widgets
import time

class Tablero:
  def __init__(self, tamano_celda=(50, 50), n_celdas=(5,5)):
    self.out = widgets.HTML()
    display(self.out)
    self.tamano_celda = tamano_celda
    self.n_celdas = n_celdas

  def dibujar(self, objetos):
    tablero = "<table border='1' >{}</table>"
    filas = ""

    for i in range(self.n_celdas[0]):
      s = ""
      for j in range(self.n_celdas[1]):
        contenido =""
        for o in objetos:
          if o.x == j and o.y == i:
            contenido = \
            "<div style='transform: rotate({angulo}deg);font-size:{tamano_emoticon}px;'>{emoticon}</div>".\
            format(angulo = o.angulo, tamano_emoticon = o.tamano_emoticon, emoticon = o.emoticon)
        s += "<td style='height:{alto}px;width:{ancho}px'>{contenido}</td>".\
          format(alto = self.tamano_celda[0], ancho = self.tamano_celda[1],
                contenido = contenido)
      filas += "<tr>{}</tr>".format(s)
    tablero = tablero.format(filas)
    self.out.value = tablero
```

```
class Aspiradora(Agente):
  def __init__(self, x=0, y=0, angulo=0, emoticon="ü§ñ", tamano_emoticon=30, energia = 5):
    super().__init__(x, y, angulo, emoticon, tamano_emoticon)
    self.energia = energia
    self.puntaje = 0

  #Movimiento
  def abajo(self):
    if self.y < 4:
      self.y += 1
      self.energia -=1

  def derecha(self):
    if self.x < 4:
      self.x += 1
      self.energia -=1

  def arriba(self):
    if self.y > 0:
      self.y -= 1
      self.energia -=1

  def izquierda(self):
    if self.x > 0:
      self.x -= 1
      self.energia -=1


  def aspirar(self, hoja, objetos):
    if self.x == hoja.x and self.y == hoja.y:
      objetos.remove(hoja)
      self.puntaje += 1

  def escogerMovimiento(self):
  # Esquinas
    if self.x == 0 and self.y == 0:
      return random.choice(["derecha", "abajo"])
    elif self.x == 4 and self.y == 0:
      return random.choice(["izquierda", "abajo"])
    elif self.x == 0 and self.y == 4:
      return random.choice(["derecha", "arriba"])
    elif self.x == 4 and self.y == 4:
      return random.choice(["izquierda", "arriba"])

    # Laterales
    elif self.x == 0 and 0 < self.y < 4:
      return random.choice(["derecha", "abajo", "izquierda"])
    elif self.y == 0 and self.x < 0 and self.x < 4:
      return random.choice(["derecha", "abajo", "arriba"])
    elif self.x == 4 and 0 < self.y < 4:
      return random.choice(["izquierda", "derecha", "arriba"])
    elif self.y ==4  and self.x > 0 and self.x < 4:
      return random.choice(["izquierda", "abajo", "arriba"])

    # Centro (permite moverse en todas las direcciones)
    else:
      return random.choice(["izquierda", "derecha", "abajo", "arriba"])

  def mover(self):
    direccion_elegida = self.escogerMovimiento()

    if direccion_elegida == "abajo":
      self.abajo()
    elif direccion_elegida == "derecha":
      self.derecha()
    elif direccion_elegida == "arriba":
      self.arriba()
    elif direccion_elegida == "izquierda":
      self.izquierda()

  #Revisar hojas
  def percibirHojas(self, hojas):
    cuadros_adyacentes = []

    # Coordenadas de los cuadros adyacentes (Norte, Sur, Oeste, Este)
    adyacentes_coords = [(self.x, self.y - 1), (self.x, self.y + 1),
                         (self.x - 1, self.y), (self.x + 1, self.y)]

    for hoja in hojas:
        if (hoja.x,hoja.y) in adyacentes_coords:
            cuadros_adyacentes.append(hoja)

    return cuadros_adyacentes


  def accion(self, objetos):
    env = self.percibirHojas(objetos)
    hojas = self.percibirHojas(env)
    if hojas:
      hoja_destino = random.choice(hojas)
      x_destino, y_destino = hoja_destino.x, hoja_destino.y

      if x_destino < self.x:
          self.izquierda()
          self.aspirar(hoja_destino, objetos)
      elif x_destino > self.x:
          self.derecha()
          self.aspirar(hoja_destino, objetos)
      elif y_destino < self.y:
          self.arriba()
          self.aspirar(hoja_destino, objetos)
      elif y_destino > self.y:
          self.abajo()
          self.aspirar(hoja_destino, objetos)
    else:
      self.mover()
```

```
class AspiradoraV2(Aspiradora):
  def __init__(self, x=0, y=0, angulo=0, emoticon="ü§ñ", tamano_emoticon=30,
               energia = 5, estado = "centro"):
    super().__init__(x, y, angulo, emoticon, tamano_emoticon, energia)
    self.estado = estado
    self.energia = energia
    self.puntaje = 0

  #Movimiento
  def abajo(self):
    if self.y < 9:
      self.y += 1
      self.energia -=1

  def derecha(self):
    if self.x < 9:
      self.x += 1
      self.energia -=1

  def arriba(self):
    if self.y > 0:
      self.y -= 1
      self.energia -=1

  def izquierda(self):
    if self.x > 0:
      self.x -= 1
      self.energia -=1

  def diagonal_superior_izquierda(self):
    if self.x > 0 and self.y > 0:
        self.x -= 1
        self.y -= 1
        self.energia -=1

  def diagonal_superior_derecha(self):
    if self.x < 9 and self.y > 0:
        self.x += 1
        self.y -= 1
        self.energia -=1

  def diagonal_inferior_izquierda(self):
    if self.x > 0 and self.y < 9:
        self.x -= 1
        self.y += 1
        self.energia -=1

  def diagonal_inferior_derecha(self):
    if self.x < 9 and self.y < 9:
        self.x += 1
        self.y += 1
        self.energia -=1


  def aspirar(self, hoja, objetos):
    if self.x == hoja.x and self.y == hoja.y:
      objetos.remove(hoja)
      self.puntaje += 1

  def mover(self):
    if self.estado == "centro":
      self.mover_al_centro()
    elif self.estado == "aleatorio":
      direccion_elegida = self.escogerMovimiento()

      direcciones_validas = {
          "abajo": self.abajo,
          "derecha": self.derecha,
          "diagonal_inferior_izquierda": self.diagonal_inferior_izquierda,
          "diagonal_inferior_derecha": self.diagonal_inferior_derecha,
          "diagonal_superior_izquierda": self.diagonal_superior_izquierda,
          "diagonal_superior_derecha": self.diagonal_superior_derecha,
          "arriba": self.arriba,
          "izquierda": self.izquierda
      }

      if direccion_elegida in direcciones_validas:
          movimiento_seleccionado = direcciones_validas[direccion_elegida]
          movimiento_seleccionado()

  def mover_al_centro(self):
      centro_x = 4
      centro_y = 4

      if self.x == 4 and self.y == 4:
        self.estado = "aleatorio"

      if self.x < centro_x:
          if self.y > centro_y:
              self.diagonal_superior_derecha()
          elif self.y < centro_y:
              self.diagonal_inferior_derecha()
          else:
              self.derecha()
      elif self.x > centro_x:
          if self.y > centro_y:
              self.diagonal_superior_izquierda()
          elif self.y < centro_y:
              self.diagonal_inferior_izquierda()
          else:
              self.izquierda()
      else:
          if self.y > centro_y:
              self.arriba()
          elif self.y < centro_y:
              self.abajo()

  def escogerMovimiento(self):
  # Esquinas
    if self.x == 0 and self.y == 0:
      self.estado = "centro"
      return random.choice(["derecha", "abajo","diagonal_inferior_derecha"])
    elif self.x == 9 and self.y == 0:
      self.estado = "centro"
      return random.choice(["izquierda", "abajo","diagonal_inferior_izquierda"])
    elif self.x == 0 and self.y == 9:
      self.estado = "centro"
      return random.choice(["derecha", "arriba","diagonal_superior_derecha"])
    elif self.x == 9 and self.y == 9:
      self.estado = "centro"
      return random.choice(["izquierda", "arriba","diagonal_superior_izquierda"])

    # Laterales
    elif self.x == 0 and 0 < self.y < 9:
      self.estado = "centro"
      return random.choice(["derecha", "abajo", "izquierda","diagonal_superior_derecha","diagonal_inferior_derecha"])
    elif self.y == 0 and self.x < 0 and self.x < 9:
      self.estado = "centro"
      return random.choice(["derecha", "abajo", "arriba","diagonal_inferior_derecha","diagonal_inferior_izquierda"])
    elif self.x == 9 and 0 < self.y < 9:
      self.estado = "centro"
      return random.choice(["izquierda", "derecha", "arriba","diagonal_superior_derecha","diagonal_superior_izquierda"])
    elif self.y == 9  and self.x > 0 and self.x < 9:
      self.estado = "centro"
      return random.choice(["izquierda", "abajo", "arriba","diagonal_inferior_izquierda","diagonal_superior_izquierda"])

    # Centro (permite moverse en todas las direcciones)
    else:
      return random.choice(["izquierda", "derecha", "abajo", "arriba",
                            "diagonal_inferior_izquierda","diagonal_inferior_derecha",
                            "diagonal_superior_izquierda","diagonal_superior_derecha"])

  #Revisar hojas
  def percibirHojas(self, hojas):
    cuadros_adyacentes = []

    # Coordenadas de los cuadros adyacentes (Norte, Sur, Oeste, Este, diagonales)
    adyacentes_coords = [(self.x, self.y - 1), (self.x, self.y + 1),
                         (self.x - 1, self.y), (self.x + 1, self.y),
                         (self.x - 1 ,self.y + 1), (self.x + 1, self.y + 1 ),
                         (self.x - 1 , self.y -1), (self.x + 1, self.y -1)]

    for hoja in hojas:
        if (hoja.x,hoja.y) in adyacentes_coords:
            cuadros_adyacentes.append(hoja)

    return cuadros_adyacentes


  def accion(self, objetos):
    env = self.percibirHojas(objetos)
    hojas = self.percibirHojas(env)
    if hojas:
      hoja_destino = random.choice(hojas)
      x_destino, y_destino = hoja_destino.x, hoja_destino.y

      if x_destino > self.x and y_destino < self.y :
          self.diagonal_superior_derecha()
          self.aspirar(hoja_destino, objetos)

      elif x_destino < self.x and y_destino > self.y :
          self.diagonal_inferior_izquierda()
          self.aspirar(hoja_destino, objetos)

      elif x_destino < self.x and y_destino < self.y :
          self.diagonal_superior_izquierda()
          self.aspirar(hoja_destino, objetos)

      elif x_destino > self.x and y_destino > self.y :
          self.diagonal_inferior_derecha()
          self.aspirar(hoja_destino, objetos)

      elif x_destino < self.x:
          self.izquierda()
          self.aspirar(hoja_destino, objetos)

      elif x_destino > self.x:
          self.derecha()
          self.aspirar(hoja_destino, objetos)
      elif y_destino < self.y:
          self.arriba()
          self.aspirar(hoja_destino, objetos)
      elif y_destino > self.y:
          self.abajo()
          self.aspirar(hoja_destino, objetos)
    else:
      self.mover()
```
