En este caso contamos con un conjunto de datos con información sobre el último viaje del trasatlántico. La información  con la que disponemos es la siguiente:

* `survived` Sobrevió ($0$ = NO, $1$ = SI)
* `pclass:` Tipo de tiquete (1ra-clase, 2da-clase, 3ra-clase)
* `sex:` Sexo
* `age:` Edad
* `sibsp:` Número de hermanos/cónyuges en el barco
* `parch:` Número de padres/hijos en el barco
* `fare:` Costo del viaje para el pasajero (en libras)
* `embarked:` Puerto de embarque (C = Cherbourg, Q = Queenstown, S = Southampton)
* `class:`Tipo de tiquete (1ra-clase, 2da-clase, 3ra-clase)
* `who:` hombre, mujer o niño.
A continuación se presentan los datos disponibles para el análisis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset ('titanic')
data.head()
```

__1.__ Identifique las variables **numéricas, ordinales y nominales**. Realice un análisis descriptivo para cada una. **En cada caso debe presentar un valor numérico y una gráfica.**

La identificación de variables numéricas, ordinales y nominales se basa en las características intrínsecas de los datos y el contexto del problema que se está abordando. A continuación, justificaré la clasificación de las variables en el DataFrame proporcionado:

1. **Variables Numéricas:**
   - `age`, `sibsp`, `parch`, y `fare`: Todas estas variables son numéricas porque representan valores continuos o discretos, como la edad, el número de hermanos/cónyuges a bordo, el número de padres/hijos a bordo y la tarifa del billete, respectivamente.

2. **Variables Ordinales:**
   - `pclass`: Aunque se almacena como un número, esta variable es ordinal, ya que representa un orden específico de clases socioeconómicas. La clase 1 es superior a la clase 2, y así sucesivamente.

3. **Variables Nominales:**
   - `sex`: Esta variable es nominal, ya que representa el género de una persona y no tiene un orden inherente.
   - `embarked`: También es una variable nominal, ya que representa el puerto de embarque y no implica ningún orden particular.
   - `class`: Aunque contiene números, esta variable se considera nominal porque representa las clases de billete en términos de texto ('First', 'Second', 'Third').
   - `who`: Representa la categoría de la persona ('man', 'woman'), que es nominal ya que no hay un orden específico.
   - `deck`: A pesar de que contiene categorías representadas por letras ('A', 'B', 'C', etc.), se considera nominal porque las letras no tienen un orden inherente.
   - `embark_town` y `alive`: Ambas son variables nominales, ya que representan el nombre del puerto de embarque y el estado de supervivencia respectivamente, sin un orden inherente.
   - `alone` y `adult_male`: Aunque contienen valores booleanos (True/False), se considera nominal porque representa si una persona viaja sola o no y si es un hombre adulto o no, respectivamente, sin un orden específico.
   - `survived`: Esta variable es nominal porque representa si una persona sobrevivió (1) o no (0).



```python
# Análisis descriptivo de variables numéricas
numeric_desc = data[['age', 'sibsp', 'parch', 'fare']].describe()

# Gráficas para variables numéricas
fig, axs = plt.subplots(2, 2, figsize=(12, 5))
axs = axs.flatten()

for i, column in enumerate(['age', 'sibsp', 'parch', 'fare']):
    sns.histplot(data[column], kde=True, ax=axs[i])
    axs[i].set_title(f'Distribución de {column}')
    axs[i].set_xlabel(column)
    axs[i].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()

print("\n Análisis Descriptivo de Variables Numéricas:")
print(numeric_desc)
```

```python
# Análisis descriptivo de la variable ordinal
ordinal_desc = data['pclass'].value_counts().sort_index()

# Gráfico para la variable ordinal
plt.figure(figsize=(5, 3))
sns.countplot(data=data, x='pclass', order=[1, 2, 3])
plt.title('Frecuencia de pclass')
plt.xlabel('pclass')
plt.ylabel('Frecuencia')
plt.show()

print("\n Frecuencia de la Variable Ordinal 'pclass':")
print(ordinal_desc)
```

```python
nominal_vars = data[['sex', 'embarked', 'class', 'who', 'deck', 'embark_town', 'alive', 'alone', 'adult_male']]

# Análisis descriptivo de frecuencia y gráficos de barras para las variables nominales
for column in nominal_vars.columns:
    # Gráfico de barras
    plt.figure(figsize=(5, 3))
    sns.countplot(data=data, x=column, order=data[column].value_counts().index)
    plt.title(f'Frecuencia de {column}')
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.show()

    # Análisis descriptivo de frecuencia
    freq_desc = data[column].value_counts()
    print(f"Frecuencia de la Variable Nominal '{column}':")
    print(freq_desc)
```

¿Cuál es la proporción de pasajeros que sobrevivió según la clase del tiquete?

```python
# Proporción de sobrevivientes por clase de tiquete
survival_by_class = data.groupby('pclass')['survived'].mean()
print("Proporción de sobrevivientes según la clase del tiquete:")
print(survival_by_class)

# Gráfico de proporción de sobrevivientes por clase
sns.countplot(data=data, x='pclass', hue='survived')
plt.xlabel('Clase del Tiquete')
plt.ylabel('Cantidad de Pasajeros')
plt.title('Proporción de Sobrevivientes por Clase de Tiquete')
plt.legend(title='Sobrevivió', labels=['No', 'Sí'])
plt.show()
```

**Interpretación**: El análisis muestra la proporción de sobrevivientes para cada clase de tiquete. Se observa que la clase de tiquete de primera clase (0.62) tiene la mayor proporción de sobrevivientes, seguida por la segunda clase (0.47) y la tercera clase (0.24).

¿Cuál es el promedio de las edades de los menores de edad que sobrevivieron en comparación con los fallecidos?


```python
# Definir menores de edad como menores de 18 años
menores_de_edad = data[data['age'] < 18]

# Promedio de edades de menores de edad sobrevivientes y fallecidos
mean_age_survived = menores_de_edad[menores_de_edad['survived'] == 1]['age'].mean()
mean_age_not_survived = menores_de_edad[menores_de_edad['survived'] == 0]['age'].mean()

print("Promedio de edades de menores de edad sobrevivientes:", mean_age_survived)
print("Promedio de edades de menores de edad fallecidos:", mean_age_not_survived)

# Gráfico de promedio de edades para menores de edad
sns.catplot(data=menores_de_edad, x='survived', y='age', kind='bar')
plt.xlabel('Sobrevivió')
plt.ylabel('Edad Promedio')
plt.title('Edad Promedio de Menores de Edad Sobrevivientes vs. No Sobrevivientes')
plt.xticks([0, 1], ['No', 'Sí'])
plt.show()
```

**Interpretación:** El promedio de edades de los menores de edad que sobrevivieron es  7.8 años, mientras que el promedio de edades de los menores de edad que fallecieron es 10.3 años.

¿Cuál fue la proporción de mujeres sobrevivientes según la clase del tiquete?

```python
# Proporción de mujeres sobrevivientes por clase de tiquete
female_survival_by_class = data[data['sex'] == 'female'].groupby('pclass')['survived'].mean()
print("Proporción de mujeres sobrevivientes según la clase del tiquete:")
print(female_survival_by_class)

# Gráfico de proporción de mujeres sobrevivientes por clase
female_survival_by_class = female_survival_by_class.reset_index()  # Convertir la Serie en un DataFrame
sns.barplot(data=female_survival_by_class, x='pclass', y='survived')
plt.xlabel('Clase del Tiquete')
plt.ylabel('Proporción de Mujeres Sobrevivientes')
plt.title('Proporción de Mujeres Sobrevivientes por Clase de Tiquete')
plt.show()
```

Interpretación: El análisis muestra la proporción de mujeres sobrevivientes para cada clase de tiquete. Se observa que la clase de tiquete de primera clase (0.96) tiene la mayor proporción de mujeres sobrevivientes, seguida por la segunda clase (0.92) y la tercera clase (0.5).

¿Cuál es el promedio de las edades de los adultos que sobrevivieron en comparación con los fallecidos?

```python
# Definir adultos como mayores de 18 años
adultos = data[data['age'] >= 18]

# Promedio de edades de adultos sobrevivientes y fallecidos
mean_age_adult_survived = adultos[adultos['survived'] == 1]['age'].mean()
mean_age_adult_not_survived = adultos[adultos['survived'] == 0]['age'].mean()

print("Promedio de edades de adultos sobrevivientes:", mean_age_adult_survived)
print("Promedio de edades de adultos fallecidos:", mean_age_adult_not_survived)

# Gráfico de promedio de edades para adultos
sns.catplot(data=adultos, x='survived', y='age', kind='bar')
plt.xlabel('Sobrevivió')
plt.ylabel('Edad Promedio')
plt.title('Edad Promedio de Adultos Sobrevivientes vs. No Sobrevivientes')
plt.xticks([0, 1], ['No', 'Sí'])
plt.show()
```

**Inteerpretación**: : El promedio de edades de los adultos que sobrevivieron es 33.79, mientras que el promedio de edades de los adultos que fallecieron es 33.45.

¿Los menores de edad de primera clase sobrevivieron en mayor proporción que los de tercera clase?

```python
# Proporción de sobrevivientes entre menores de edad de primera y tercera clase
survival_by_class_and_age = data[(data['age'] < 18) & ((data['class'] == 'First') | (data['class'] == 'Third'))].groupby(['class'])['survived'].mean()
print("Proporción de sobrevivientes entre menores de edad de primera y tercera clase:")
print(survival_by_class_and_age)

# Gráfico de proporción de sobrevivientes entre menores de edad de primera y tercera clase
survival_by_class_and_age = survival_by_class_and_age.reset_index()  # Convertir la Serie en un DataFrame
sns.barplot(data=survival_by_class_and_age, x='class', y='survived')
plt.xlabel('Clase del Tiquete')
plt.ylabel('Proporción de Sobrevivientes')
plt.title('Comparación de Proporción de Sobrevivientes entre Menores de Edad de 1ra y 3ra Clase')
plt.show()
```

Interpretación: La gráfica muestra la proporción de sobrevivientes entre los menores de edad de primera y tercera clase. Se observa que la proporción de sobrevivientes entre los menores de edad de primera clase (0.91) es mayor que la proporción de sobrevivientes entre los menores de edad de tercera clase (0.37).

Existe diferencia entre la proporción de mujeres sobrevivientes de tercera clase que sobrevivieron  y embarcaron en Cherbourg

```python
# Proporción de mujeres sobrevivientes de tercera clase que embarcaron en Cherbourg
female_survival_by_class_and_embarked = data[(data['sex'] == 'female') & (data['class'] == 'Third') & (data['embarked'] == 'C')]['survived'].mean()

# Gráfico de barra
plt.figure(figsize=(6, 4))
sns.barplot(x=['Mujeres de 3ra Clase (Cherbourg)'], y=[female_survival_by_class_and_embarked])
plt.ylabel('Proporción de Sobrevivientes')
plt.title('Proporción de Mujeres de 3ra Clase de Cherbourg Sobrevivientes')
plt.ylim(0, 1)  # Establecer el rango del eje y para mostrar proporciones
plt.show()

# Imprimir la proporción numérica
print("Proporción de mujeres sobrevivientes de tercera clase que embarcaron en Cherbourg:", female_survival_by_class_and_embarked)
```

**Interpretación:** La proporción de mujeres sobrevivientes de tercera clase que embarcaron en Cherbourg es un 0.65, que podemos ver en la gráfica anterior. Esta proporción representa la cantidad de mujeres de tercera clase que sobrevivieron y embarcaron en Cherbourg en relación con el total de mujeres de tercera clase en el Titanic.
