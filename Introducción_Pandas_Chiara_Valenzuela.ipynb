# __Series__


```python
import numpy as np
import pandas as pd
```

## __Creando series__

Declare una serie vacia

```python
serie_vacia = pd.Series()

print(serie_vacia)
```

Dada la lista `X = ['A','B','C', 'D']` conviertala a un serie de Pandas




```python
X = ['A', 'B', 'C', 'D']
serie_x = pd.Series(X)
print(serie_x)
```

Asignele el nombre  `'Letras'` a la serie `X`

```python
serie_x = pd.Series(X, name='Letras')
print(serie_x)
```

Muestre los valores de la serie `X`

```python
print(serie_x.values)
```

## __Indexaci√≥n__

Asigne los √≠ndices `[primero, segundo, tercero, cuarto]` a la serie `X`

```python
nuevos_indices = ['primero', 'segundo', 'tercero', 'cuarto']
serie_x = pd.Series(X, index=nuevos_indices)
print(serie_x)
```

Imprima el primer elemento de la serie `X` (utilice diferentes m√©todos)



```python
primer_elemento = serie_x.iloc[0]
print(primer_elemento)

primer_elemento = serie_x.loc['primero']
print(primer_elemento)

primer_elemento = serie_x[0]
print(primer_elemento)
```

Imprima el primer y el √∫ltimo de la serie `X` (utilice diferentes m√©todos)

```python
primer_elemento, ultimo_elemento = serie_x.iloc[0], serie_x.iloc[-1]
print("Primer elemento:", primer_elemento, "√öltimo elemento", ultimo_elemento)

primer_elemento, ultimo_elemento = serie_x.loc['primero'], serie_x.loc['cuarto']
print("Primer elemento:", primer_elemento, "√öltimo elemento", ultimo_elemento)

primer_elemento, ultimo_elemento  = serie_x[0], serie_x[-1]
print("Primer elemento:", primer_elemento, "√öltimo elemento", ultimo_elemento)

```

Imprima todos los elementos intermedios de la serie `X` (utilice diferentes m√©todos)

```python
elementos_intermedios = serie_x.iloc[1:-1]
print("Elementos intermedios:", elementos_intermedios)

elementos_intermedios = serie_x.loc['segundo':'tercero']
print("Elementos intermedios:", elementos_intermedios)

elementos_intermedios = serie_x[1:-1]
print("Elementos intermedios:", elementos_intermedios)
```

Imprima todos los elementos de la serie `X` en orden inverso (utilice diferentes m√©todos)

```python
elementos_inversos = serie_x.iloc[::-1]
print("Elementos en orden inverso:", elementos_inversos)

elementos_inversos = serie_x.loc[['cuarto', 'tercero', 'segundo', 'primero']]
print("Elementos en orden inverso:", elementos_inversos)

elementos_inversos = pd.Series(reversed(serie_x))
print("Elementos en orden inverso:", elementos_inversos)
```
